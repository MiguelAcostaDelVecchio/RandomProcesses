```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
from IPython import display
```

```python
# Hand Calculation
# I tried uploading pdf/images to jupyter notebook without luck
# Hand calculation is shown in my github: https://github.com/MiguelAcostaDelVecchio/RandomProcesses
# Work is shown in Week3/HW3_HandCalc.pdf

# Question 1 calculation to get graph
p_4 = []

while len(p_4)<5000:
    p = stats.uniform.rvs()
    h = stats.binom(10,p).rvs()
    if h==4:
        p_4.append(p)
```

```python
# Question 1 plot graph
plt.hist(p_4 , bins=25  , density=True , label="simulated" )
p = np.linspace(0,1)
pdf = 2310 * p**4 * (1-p)**6
plt.plot(p,pdf , label="bayes thm")
plt.legend()
plt.xlabel("p")
plt.ylabel("pdf(p)")
plt.title("Question1: pdf(p | 4 heads) assuming uniform prior")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
# generate a value of p at random
# use it to flip a coin 5x
# observe to see if exactly three of those are heads
# if yes, save that value of p in a list
# repeat all prior steps many times
# wind up with a list of values of p that all suceeded in getting exactly 3 heads
# look a the distribution of those values of p

p_3 = []

while len(p_3)<5000:
    p = stats.uniform.rvs()
    h = stats.binom(5,p).rvs()
    if h==3:
        p_3.append(p)
```

```python
plt.hist(p_3 , bins=25  , density=True , label="simulated" )
p = np.linspace(0,1)
pdf = 60 * p**3 * (1-p)**2
plt.plot(p,pdf , label="bayes thm")
plt.legend()
plt.xlabel("p")
plt.ylabel("pdf(p)")
plt.title("pdf(p | 3 heads) assuming uniform prior")
plt.show()
```

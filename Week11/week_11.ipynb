```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
import scipy.signal as signal
```

```python
n = np.arange(1000)
x = np.cos(n * np.pi/50)
noise = stats.norm(loc=0,scale=1).rvs(1000)
y = x+noise
ryy = signal.correlate(y,y)
# plt.plot(n,y,'o')
plt.plot(ryy,'o')
plt.xlim([800,1200])
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
# sum of binomial rv's is well modeled by a normal

h = np.arange(1100,1400)
pr_b = stats.binom(p=0.5,n=2500).pmf(h)               # binomial (true)
pr_n = stats.norm(loc=1250 , scale=25).pdf(h[::5])    # normal (estimate)
plt.plot(h      ,pr_b       , label ='binomial')
plt.plot(h[::5] ,pr_n , 'o' , label='normal')
plt.legend()
plt.show()

pr_2000 = 1 - stats.norm(loc=1250,scale=25).cdf(2000)
print(f"{pr_2000:0.8f}")
```

```python
# sum of uniform rv's is well modeled by a normal

vals = stats.uniform(loc = 1 , scale=4).rvs((10000 , 250)).sum(axis=1)
v = np.linspace(700,800,num=250)
p = stats.norm(loc=750 , scale=np.sqrt(1000/3)).pdf(v)
plt.hist(vals , density=True , bins=25)
plt.plot(v,p)
plt.show()
print(f"{vals.mean():0.2f}")
print(f"{vals.var():0.2f}")
```

```python
# average of 250 uniform rv's is well modeled by a normal

vals = stats.uniform(loc = 1 , scale=4).rvs((10000 , 250)).mean(axis=1)
v = np.linspace(2.5,3.5,num=250)
p = stats.norm(loc=3 , scale=np.sqrt(4/750)).pdf(v)
plt.hist(vals , density=True , bins=25)
plt.plot(v,p)
plt.show()
print(f"{vals.mean():0.2f}")
print(f"{vals.var():0.2f}")
```

```python
vals = stats.uniform(loc = 1 , scale=4).rvs((100,5)).sum(axis=1)
print(vals.mean())
print(vals.std())
```

```python
n = 1
while True:
    mu  = 0.5
    var = (0.5*0.5)/n
    std = np.sqrt(var)
    cdf = stats.norm(loc=mu,scale=std).cdf(0.625)
    print(n,f"{cdf:0.4f}")
    if cdf > 0.95:
        break
    else:
        n = n + 1

# so if n>=44 then there is at least an 95% we will correctly identify it as the p=0.5 coin
```

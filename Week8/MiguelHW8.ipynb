```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
from numpy.random import seed
from numpy.random import normal
```

## Step 2

```python
#seed(1)
# There are a total of 260 business days in 2022
businessDays = 260

# Number of simulations
nSim = 10000

# yList[0] = Y1 = Price of stock in first business day of the year
yList = [30]

# Assume Xn is a normal distribution with mean = 0 and var = 0.25
v = np.linspace(0,businessDays,num=businessDays)
#xList = stats.norm(loc=0 , scale=0.25).pdf(v)
xList = normal(loc = 0, scale = np.sqrt(0.25), size = businessDays)
```

```python
plt.plot(v,xList)
plt.xlabel("Business day in 2022")
plt.ylabel("Price change of stock")
plt.title("Price change of stock")
plt.show()
```

```python
for i in range(0,businessDays):
    yList.append(xList[i] + yList[i])
```

```python
v1 = np.linspace(0, businessDays+1,num=businessDays+1)
plt.plot(v1,yList)
plt.xlabel("Business day in 2022")
plt.ylabel("Stock price")
plt.title("Stock price every business day of 2022")
plt.show()
```

```python
xListN = normal(loc = 0, scale = np.sqrt(0.25), size = (nSim, businessDays))
```

```python
aMin = -5
aMax = 5

cnt1 = 0
cnt2 = 0
cnt3 = 0

for i in range(0,nSim):
    if xListN[i][:2].sum() < aMax and xListN[i][:2].sum() > aMin:
        cnt1 += 1
    #else:
    #    print(f"Y2: {xListN[i][:2].sum():0.2f}")

    if xListN[i][:11].sum() < aMax and xListN[i][:11].sum() > aMin:
        cnt2 += 1
    #else:
    #    print(f"Y11: {xListN[i][:11].sum():0.2f}")

    if xListN[i][:101].sum() < aMax and xListN[i][:101].sum() > aMin:
        cnt3 += 1
    #else:
    #    print(f"Y101: {xListN[i][:101].sum():0.2f}")

"""
print(cnt1)
print(cnt2)
print(cnt3)
"""

print(f"The probability that Y2 is between 25 and 35 is {(cnt1/nSim)*100:0.2f}%")
print(f"The probability that Y11 is between 25 and 35 is {(cnt2/nSim)*100:0.2f}%")
print(f"The probability that Y101 is between 25 and 35 is {(cnt3/nSim)*100:0.2f}%")
```

## Step 1

```python
a, b = -100, 100
mu, sigma = 0, np.sqrt(0.25)
dist = stats.truncnorm((a - mu) / sigma, (b - mu) / sigma, loc=mu, scale=sigma)

values = []
for i in range(0,nSim):
    values.append(dist.rvs(businessDays))
```

```python
cnt3 = 0
cnt4 = 0
cnt5 = 0

for i in range(0,nSim):
    if values[i][:2].sum() < aMax and values[i][:2].sum() > aMin:
        cnt3 += 1
    else:
        print(f"Y2: {values[i][:2].sum():0.2f}")

    if values[i][:11].sum() < aMax and values[i][:11].sum() > aMin:
        cnt4 += 1
    #else:
    #    print(f"Y11: {values[i][:11].sum():0.2f}")

    if values[i][:101].sum() < aMax and values[i][:101].sum() > aMin:
        cnt5 += 1
    #else:
    #    print(f"Y101: {values[i][:101].sum():0.2f}")


#print(cnt3)
#print(cnt4)
#print(cnt5)


print(f"The probability that Y2 is between 25 and 35 is {(cnt3/nSim)*100:0.2f}%")
print(f"The probability that Y11 is between 25 and 35 is {(cnt4/nSim)*100:0.2f}%")
print(f"The probability that Y101 is between 25 and 35 is {(cnt5/nSim)*100:0.2f}%")
```

The results from Step 1 and Step 2 proof the theory of the central limit theorem. Everything
can be estimated using a normal distribution and it will yield relatively good results most
of the time.
